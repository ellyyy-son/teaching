```python
import pandas as pd
```

# Load dataset

```python
data = pd.read_excel("data_transformation.xlsx", index_col="Instance")
data.head()
```

# Label Encoding

```python
yes_no_columns = [
  'Alternative', 'Bar', 'Friday&Saturday', 'Hungry', 'Raining', 'Reservation',
  'WillWait'
]
categorical_columns = [
  'Patrons', 'Price Range', 'TypeCuisine', 'EstimatedWaitingTime'
]
```

## Label Encoding

```python
# convert yes_no_columns to boolean
from sklearn.preprocessing import LabelEncoder

transformed_data = data.copy()

le = LabelEncoder()
for column in yes_no_columns:
  transformed_data[column] = le.fit_transform(
    data[column]
  )

transformed_data[yes_no_columns]
```

## One Hot Encoding

```python
# convert categorical columns to one-hot encoding
from sklearn.preprocessing import OneHotEncoder

# one_hot_encoders = []

for column in categorical_columns:
  one_hot_encoder = OneHotEncoder(handle_unknown='ignore')
  one_hot_encoded = one_hot_encoder.fit_transform(
    data[column].values.reshape(-1, 1)
  )
  # one_hot_encoders.append((column, one_hot_encoder))

  one_hot_encoded_df = pd.DataFrame(
    one_hot_encoded.toarray(),
    columns=[f"{column}_{category}" for category in one_hot_encoder.categories_[0]]
  ).astype(int)


one_hot_encoded_df
```
